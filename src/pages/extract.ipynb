```python
from bs4 import *
import re
import requests

name='Stephen Curry'
url = "https://en.wikipedia.org/wiki/"+name.replace(' ', '_')
r = requests.get(url)
td=[]
th=[]
if r.status_code == 200:
    # print('Found '+name+' on Wikipedia')

    # Get body content
    soup = BeautifulSoup(r.text, 'html.parser').select('body')[0]
    # print(soup)
    req = ''

    # Iterate throught all tags
    for tag in soup.find_all():
        if tag.name == "td":
            # print(tag.text)
            td.append(tag.text)
            
        if tag.name == "th":
            # print(tag.text)
            th.append(tag.text)
th.pop(0)
td.pop(0)
lst=[]

for tag in soup.find_all():
        tag = soup.th
        attribute = tag['class']
        # print(attribute[0])
        if attribute[0] == 'infobox-header':
            new=tag.get_text('class')
            if new not in lst:
                lst.append(new)
                
print(lst)
dict={}

print(th.index('\n'))
index=th.index('\n')


for i in range(index-1):
    dict[th[i]]=td[i]

[print(key,':',value) for key, value in dict.items()]
```
